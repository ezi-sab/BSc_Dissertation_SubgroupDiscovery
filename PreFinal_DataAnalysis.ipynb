### DataFrame without brown genes

i.e. Add frequency of brown gene into root and make one more columns to count frequency of brown genes.

```python
total_genes_with_brown = []


for key in serotypes:
    # Simplify the sequence of genes
    simply_genes = simplify_genes(serotypes[key]['genes']) #list
    
    # Add gene as one element in list "total_genes_with_brown"
    for i in range(len(simply_genes)):
        if simply_genes[i] in total_genes_with_brown:
            continue # to avoid duplicate
        else:
            total_genes_with_brown.append(simply_genes[i])
#len(total_genes_with_brown) = 193
```

```python
brown_genes = []

for i in range(len(total_genes_with_brown)):
    if '-' in total_genes_with_brown[i]:
        brown_genes.append(total_genes_with_brown[i])
#len(brown_genes) = 35
```

```python
total_genes = [] #Store all genes in dataset without '-'
       
for i in range(len(total_genes_with_brown)):
    if total_genes_with_brown[i] in brown_genes:
        remove_dash = total_genes_with_brown[i][:-1]
        
        if remove_dash in total_genes:
            continue
        else:
            total_genes.append(remove_dash)
    
    else:
        if total_genes_with_brown[i] in total_genes:
            continue
        else:
            total_genes.append(total_genes_with_brown[i])
# len(total_genes) = 178
```

```python
# Set maximum num of rows & cols
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
fre_total_genes = dict()

# Build basic ditionary for frequency of all genes in dataset
for gene in total_genes:
    fre_total_genes[gene] = 0

for key in serotypes:
    # Simplify the sequence of genes
    simply_genes = simplify_genes(serotypes[key]['genes'])
    
    for i in range(len(simply_genes)):
        if simply_genes[i] in fre_total_genes:
            fre_total_genes[simply_genes[i]] += 1
        else:
            if '-' in simply_genes[i]:
                remove_dash = (simply_genes[i])[:-1]
                
                if remove_dash in fre_total_genes:
                    fre_total_genes[remove_dash] +=1
            else:
                fre_total_genes[simply_genes[i]] = 1
        
#fre_total_genes  
```

```python
fre_genes_as_brown = dict()
fre_genes_not_brown = dict()

ser_genes_as_brown = dict()
ser_genes_not_brown = dict()

# Build basic ditionary for frequency of all genes in dataset
for gene in total_genes:
    fre_genes_as_brown[gene] = 0
    fre_genes_not_brown[gene] = 0
    
    ser_genes_as_brown[gene] = []
    ser_genes_not_brown[gene] = []

for key in serotypes:
    # Simplify the sequence of genes
    simply_genes = simplify_genes(serotypes[key]['genes'])
    
    for i in range(len(simply_genes)):
        if '-' in simply_genes[i]:
            remove_dash = (simply_genes[i])[:-1]
                
            if remove_dash in fre_genes_as_brown:
                    fre_genes_as_brown[remove_dash] +=1
                    ser_genes_as_brown[remove_dash].append(key)
        else:
            fre_genes_not_brown[simply_genes[i]] += 1
            ser_genes_not_brown[simply_genes[i]].append(key)

#fre_genes_as_brown
#fre_genes_not_brown
```

```python
simply_genes = [] #Store gene sequences of serotype which is simplified
lin_sugars = [] #Store sugar structure of serotype

#Store pairgene-pairs of sugars in each serotypey
wholePairs_data = dict() #[key]serotypes[key] [key][key]"tuple" of pairgenes [key][key[value]List of "list" of pairsugars

for key in serotypes:
    # ignore serotypes which have no structure of sugar
    if len(serotypes[key]['sugars']) == 0:
        continue
    
    # with serotypes which have own sugar structure
    # simplify the sequence of genes
    simply_genes = simplify_genes(serotypes[key]['genes'])
    # copy the [key]['sugars'] as local 
    lin_sugars = serotypes[key]['sugars']
    
    sugars_pair = []
    for i in range(len(lin_sugars)-1): 
        li_sugars_pair = []
        li_sugars_pair.append(lin_sugars[i])
        li_sugars_pair.append(lin_sugars[i+1])
        
        sugars_pair.append(li_sugars_pair)
    
    genes_sugars_pairs = dict()
    for x in range(len(simply_genes)-1):
        genes_sugars_pairs[(simply_genes[x], simply_genes[x+1])] = sugars_pair
        
    wholePairs_data[key] = genes_sugars_pairs   
```

```python
#Store pairs of gene - pairs of sugar 
wholePairs = dict() #[key]"tuple" of pairgenes [value]List of "list" of pairsugars

for key in wholePairs_data:
    for pairgene in wholePairs_data[key]:
        for i in range(len(wholePairs_data[key][pairgene])):
            li_draft = []
            if not pairgene in wholePairs:
                wholePairs[pairgene] = []
                wholePairs[pairgene].append(wholePairs_data[key][pairgene][i])
            else:
                if wholePairs_data[key][pairgene][i] in wholePairs[pairgene]:
                    continue                
                else:
                    wholePairs[pairgene].append(wholePairs_data[key][pairgene][i])
```
